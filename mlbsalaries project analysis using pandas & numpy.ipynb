# 20. How to Sort records by player in ascending order for the year 2010?

```python
records_2010 = mlbsalaries_data[mlbsalaries_data['Year'] == 2010]

sorted_records_2010 = records_2010.sort_values(by='Player', ascending=True)

print(sorted_records_2010)
```
